```python
#The following are the pre-processing steps involved in this exercise:
#1. Lower casing: Each text is coverted to lower case
#2. Removing the URLS's : Links starting with "http" or "https" or "www" are replaced by "".
#3. Removing the Usernames: Replace @Usernames with word "". 
#4. Removing short words: Words with length less than 2 are removed
#5. Removing Stopwords : A stop word is a commonly used words(such as "the", "a", "an") that a search engine 
#                        has been programmed to ignore, both when indexing entries for searching and when retrieving
#                        them as the result of a search query.
#                        We would not want these words to take up space in database, or taking up vaulable processing time
#                        For this, we can remove them easily, by storing a list of words that consider to stop words.
#6. Lemmatizing: Lemmatization is the process of converting a word to its base form.(e.g: "wolves" to "wolf")
```

```python
urlPattern = r"((http://)[^ ]*|(https://)[^ ]*|( www\.)[^ ]*)"
userPattern = '@[^\s]+'
def process_tweets(tweet):
    #Lower casing
    tweet =  tweet.lower()
    #Removing the first character of the tweet
    tweet = tweet[1: ]
    #Removing all URL'S
    tweet = re.sub(urlPattern, '', tweet)
    #Removing all @Usernames
    tweet = re.sub(userPattern, '', tweet)
    #Removing the punctuations
    tweet = tweet.translate(str.maketrans("", "", string.punctuation))
    #tokenizing words
    tokens = word_tokenize(tweet)
    #Removing the Stop words
    final_tokens = [w for w in tokens if w not in stopwords]
    #Reducing a word to its word stem
    wordLemm = WordNetLemmatizer()
    finalwords = []
    for w in final_tokens:
        if(len(w) > 1):
            word = wordLemm.lemmatize(w)
            finalwords.append(word)
    return ' '.join(finalwords)
    
```

```python
masterData['processed_tweets'] = masterData['text'].apply(lambda x: process_tweets(x))
print('Text Preprocessing complete...')
```

```python
#Vectorizing and splitting the data
#Storing input variable-processes_tweets to X and output variable-polarity to Y
X = masterData['processed_tweets'].values
Y = masterData['polarity'].values
print(X.shape)
print(Y.shape)
```

```python
#Coverting text to word frquency vectors
#TF-IDF
#This is an acronym that stands for Term-Frequency-Inverse Documnent Freuqency which are the components of the resulting
#scores assigned to each word.
# - Term Frequency : This summarizes how often a given word appears within document.
# - Inverse Document Frequency : This downscales words that appear a lot across documents.
```

```python
#Conver a collection of raw documents to a matrix of TF-IDF features
vector = TfidfVectorizer(sublinear_tf = True)
X = vector.fit_transform(X)
print(f'Vector fitted')
print('No. of feature_words: ', len(vector.get_feature_names()))
```

```python
print(X.shape)
print(Y.shape)
```

```python
#Split train and test
#The Processed data is divided into 2 sets of data:
#1. Training Data : The dataset upon which the model would be trained on. Contains 80% data
#2. Test Data : The dataset upon whichh the model would be tested against. Contains 20% data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.20, random_state = 101)
```

```python
print("X_train", X_train.shape)
print("Y_train", Y_train.shape)
print()
print("X_test", X_test.shape)
print("Y_test", Y_test.shape)
```

```python
#Model Building
def model_evaluation(model):
    #accuracy of the model training data
    acc_train = model.score(X_train, Y_train)
    #accuracy of model on test data
    acc_test =  model.score(X_test, Y_test)
    
    print('Accuracy of model on training data: {}'.format(acc_train*100))
    print('Accuracy of model on test data: {} \n'.format(acc_test*100))
    
    #Predict values for Test dataset
    Y_pred = model.predict(X_test)
    
    #Print the evaluation metrics for the dataset
    print(classification_report(Y_test, Y_pred))
    
    #Compute and Plot the confusion matrix
    cf_matrix = confusion_matrix(Y_test, Y_pred)
    
    categories = ['Negative', 'Positive']
    group_names = ['True Neg', 'False Pos', 'False Neg', 'True Pos']
    group_percentages = ['{0:.2%}'.format(value) for value in cf_matrix.flatten()/np.sum(cf_matrix)]
    
    labels = [f'{v1}\n{v2}' for v1, v2 in zip(group_names, group_percentages)]
    labels = np.asarray(labels).reshape(2,2)
    
    sns.heatmap(cf_matrix, annot = labels, cmap = 'Reds', fmt = '', xticklabels  = categories, yticklabels = categories)
    
    plt.xlabel("Predicted values", fontdict = {'size:14'}, labelpad = 10)
    plt.ylabel("Actual values", fontdict = {'size:14'}, labelpad = 10)
    plt.xlabel("Confusion matrix", fontdict = {'size:18'}, labelpad = 20)    
```

```python
#Logistic Regression
lg = LogisticRegression()
history = lg.fit(X_train, Y_train)
model_evaluation(lg)
```

```python
#Linear SVM
svm = LinearSVC()
svm.fit(X_train, Y_train)
model_evaluation(svm)
```

```python
#Random Forest
rf = RandomForestClassifier(n_estimators = 20, criterion = 'entropy', max_depth = 50)
rf.fit(X_train, Y_train)
model_evaluation(rf)
```

```python
#Naive Bayes
nb = BernoulliNB()
nb.fit(X_train, Y_train)
model_evaluation(nb)
```

```python
#RNN
#What is RNN
#Recurrent Neural Network(RNN) are the state of the art algorithm for sequential data and are used
#by Apple's Siri and Google's voice search. This algorithm are commonly used for ordinal or temporal problems, such as 
#language translation, natural language processing(NLP), speech recognition and image processing.The algorithm is based
#on the forward propogation and backward propogation like any deep neural network.But unlike, other networks it stores
#the past outcomes and is used in the next consecutive layer with different weight.

#Embedding Layer
#Embedding layer is one of the available layers in Keras.This is mainly used in Natural Language Processing related 
#applications such as language modeling, but it can also be used with other tasks that involve neural network.While
#dealing with NLP problems, we can use pre-trained word embeddings such GloVe.Alternatively we can also train our
#own embeddings using Keras embedding layer.

#LSTM layer.
#Long Short Term Memory networks, usually called "LSTMs" were introduced by Hochreiter and Schmiduber. These have widely
#used for speech recognition, language modeling, sentiment analysis and text prediction.LSTM mainly introduced because
#of drawbacks in RNN. During RNN operation, there are 2 main issues can occur i.e. gradient vanishing and gradient exploding.
#Gradient Vanishing could results in very minute change in the weights in the backward propogation while approaching
#to local minima this may results in high run time with no convergence. On otherhand, gradient exploding results in very high 
#delta change in weights during backward propogation this may also results into no-convergence. With the LSTM these drawbacks 
#can be encounterted through 1. Memory Unit, 2. Forget Gate 3. Input layer and 4. Output layer. 

#In each iteration of the LSTM, when a new context is introduced, the past context is forgotten(not completely but partially)
#and a new context is being added to the existing context through dot operation and addition operation. 
```

```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras import regularizers

max_words = 5000
max_len = 200

tokenizer = Tokenizer(num_words = max_words)
tokenizer.fit_on_texts(masterData.processed_tweets)
sequences = tokenizer.texts_to_sequences(masterData.processed_tweets)
tweets = pad_sequences(sequences, maxlen = max_len)
print(tweets)
X_train, X_test, Y_train,Y_test = train_test_split(tweets, masterData.polarity.values, test_size = 0.2, random_state = 101)
```

```python
from keras.models import Sequential 
from keras import layers
from keras import regularizers
from keras import backend as K
from keras.callbacks import ModelCheckpoint
model2 = Sequential()
model2.add(layers.Embedding(max_words, 128))
model2.add(layers.LSTM(64, dropout = 0.5))
model2.add(layers.Dense(16, activation = 'relu'))
model2.add(layers.Dense(8, activation = 'relu'))
model2.add(layers.Dense(1, activation = 'sigmoid'))
model2.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
checkpoint2 = ModelCheckpoint("rnn_model.hdf5", monitor = 'val_accuracy', verbose = 1, save_best_only = True, 
                             mode = 'auto', save_freq = 1, save_weights_only = False)
history = model2.fit(X_train, Y_train, epochs = 10, validation_data = (X_test, Y_test), callbacks = [checkpoint2])
```

```python
sequence = tokenizer.texts_to_sequences(['this machine algorithm is the worst'])
test = pad_sequences(sequence, maxlen = max_len)
pred = model2.predict(test)
if pred > 0.5:
    print('Positive')
else:
    print('Negative')
```

```python
#Predict using saved model
def load_models():
    #Load the vectoriser
    file = open('vectoriser.pickle', 'rb')
    vectoriser = pickle.load(file)
    file.close()
    #Load the LR model
    file = open('logisticRegression.pickle', 'rb')
    lg = pickle.load(file)
    file.close()
    return vectoriser, lg
```

```python
def predict(vectoriser, model, text):
    #predict the sentiment
    processes_text = [process_tweets(sen) for sen in text]
    textdata = vectoriser.transform(processes_text)
    sentiment = model.predict(textdata)
    
    #Make a list of text with sentiment
    data = []
    for text, pred in zip(text, sentiment):
        data.append((text, pred))
    
    #Convert the list into a Pandas DataFrame
    df = pd.DataFrame(data, columns = ['text', 'sentiment'])
    df = df.replace([0, 1], ['Negative', 'Positive'])
    return df
    
```
